```python
predictors = ' + '.join(Boston_Housing.columns.difference(['CRIM']))
Multiple_Regression = smf.ols('CRIM ~ {}'.format(predictors),data = Boston_Housing).fit()
print(Multiple_Regression.summary())
Multiple_Regression.params
```

```python
import matplotlib.pyplot as plt
X = [AGE.params[1], B.params[1], CHAS.params[1],DIS.params[1],INDUS.params[1], LSTAT.params[1], MDEV.params[1], NOX.params[1], PTRATIO.params[1], RAD.params[1], RM.params[1], TAX.params[1], ZN.params[1]]
y = Multiple_Regression.params[1::1]

#Source:https://stackoverflow.com/questions/52340435/extract-regressions-coefficient-from-statsmodels
```

```python
plt.figure(figsize = (12,8))
plt.ylim(-15,15)
plt.scatter(X,y)
plt.title('Univariate vs Multivariate')
plt.xlabel("Univariate")
plt.ylabel("Multivariate")
```
